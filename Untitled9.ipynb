```python
import pandas as pd
df = pd.read_excel('supermarket.xlsx')
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
d=df[['Date','Quantity']]
print(d)
print(d.count())
#df.query("`Courses Fee` >= 23000 and `Courses Fee` <= 24000")
```

```python
#d1=d.query("`Date` >= 2019-01-01 and `Date` <= 2019-04-30")

de = d.loc[(d['Date'] >= '2019-01-01')
                     & (d['Date'] <= '2019-01-31')]
print(de)
print(de.count())
dd1=de["Quantity"].sum()
print(dd1)
```

```python
de = d.loc[(d['Date'] >= '2019-02-01')
                     & (d['Date'] <='2019-02-28')]
print(de)
print(de.count())
dd2=de["Quantity"].sum()
print(dd2)
```

```python
de = d.loc[(d['Date'] >= '2019-03-01')
                     & (d['Date'] <= '2019-03-31')]
print(de)
print(de.count())
dd3=de["Quantity"].sum()
print(dd3)
```
